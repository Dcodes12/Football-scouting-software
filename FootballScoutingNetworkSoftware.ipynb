```python
import pandas as pd
from sqlalchemy import create_engine

data = pd.read_csv('./Desktop/DataFm/fm2023.csv') #data 
data
```

```python
selected_cols = [2] + list(range(4,17)) + list(range(22,50)) + list(range(58,65)) #excluding irrelevant columns
df = data.iloc[:, selected_cols]
df
```

```python
# filtering data
df1 = pd.DataFrame(df)
df1.columns = df1.columns.str.strip()

df1[['Min Value', 'Max Value']] = df1['Transfer Value'].str.extract('€([\d.]+)M - €([\d.]+)M').astype(float)
#converting transfer value to float type

df1
```

```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
#Using young prospects as priority data
maxAge = df1['Age'] < 18 #players younger than 18
valueRange = (df1['Min Value'] > 1) #players with transfer value more than €1 million

mainDf = df1[maxAge & valueRange]

mainDf
```

```python
positionValues = mainDf['Position'].str.strip().unique()
```

```python
#panel widgets
valueSlider = pn.widgets.IntSlider(name='Max Value (millions)', start=1, end=80, step=5, value=80)
finSlider = pn.widgets.IntSlider(name='Finishing (min)', start=1, end=20, step=1, value=1)
heaSlider = pn.widgets.IntSlider(name='Heading (min)', start=1, end=20, step=1, value=1)
pacSlider = pn.widgets.IntSlider(name='Pace (min)', start=1, end=20, step=1, value=1)
pasSlider = pn.widgets.IntSlider(name='Passing (min)', start=1, end=20, step=1, value=1)
driSlider = pn.widgets.IntSlider(name='Dribbling (min)', start=1, end=20, step=1, value=1)
posSlider = pn.widgets.IntSlider(name='Positioning (min)', start=1, end=20, step=1, value=1)
hanSlider = pn.widgets.IntSlider(name='Handling (min)', start=1, end=20, step=1, value=1)
refSlider = pn.widgets.IntSlider(name='Reflexes (min)', start=1, end=20, step=1, value=1)


valueSlider
```

```python
#Radio buttons for positions
posButton = pn.widgets.RadioButtonGroup(
    name='Position Selector', 
    options=['GK (goalkeeper)','D (defender)','WB (wingback)','DM (defensive midfielder)','M (midfielder)','AM (attacking midfielder)','ST (striker)'],
    button_type='success',
    orientation='vertical'
)

posButton
```

```python
#table for dashboard
table = pn.widgets.Tabulator(mainDf, width=800, height=400)

@pn.depends(valueSlider.param.value, posButton.param.value, finSlider.param.value, heaSlider.param.value, pacSlider.param.value, pasSlider.param.value, driSlider.param.value, posSlider.param.value, hanSlider.param.value, refSlider.param.value)
def update_plot(max_value,position,fin,hea,pac,pas,dri,pos,han,ref):
    #getting main pos of selected pos
    selectedMainPos = position.split()[0]
    
    #filtered data based on user interaction
    filteredData = mainDf[(mainDf['Max Value'] <= max_value) & (mainDf['Best Pos'].str.split().str.get(0) == selectedMainPos) & 
                         (mainDf['Fin'] >= fin) & 
                         (mainDf['Hea'] >= hea) &
                         (mainDf['Pac'] >= pac) &
                         (mainDf['Pas'] >= pas) &
                         (mainDf['Dri'] >= dri) &
                         (mainDf['Pos'] >= pos) &
                         (mainDf['Han'] >= han) &
                         (mainDf['Ref'] >= ref)]
    
    #updating table
    table.value = filteredData
    
    #create and update plot
    
    #SCATTER GRAPH

    TOOLTIPS = HoverTool(tooltips=[
        ("Name", "@Name"),
        ("Age", "@Age"),
        ("Style", "@Style"),
        ("Division", "@Division"),
        ("Club", "@Club"),
        ("Nationality", "@Nat")
    ]
                        )

    p = figure(title='Players you are looking for', x_axis_label='Min Value', y_axis_label='Max Value', width=800, height=600, tools=[TOOLTIPS])
    p.circle('Min Value', 'Max Value', size=10, source=filteredData)

    return pn.Column(p, table)


@pn.depends(valueSlider.param.value, posButton.param.value, finSlider.param.value, heaSlider.param.value, pacSlider.param.value, pasSlider.param.value, driSlider.param.value, posSlider.param.value, hanSlider.param.value, refSlider.param.value)
def update_plot1(max_value,position,fin,hea,pac,pas,dri,pos,han,ref):
    #getting main pos of selected pos
    selectedMainPos = position.split()[0]
    
    #filtered data based on user interaction
    filteredData = mainDf[(mainDf['Max Value'] <= max_value) & (mainDf['Best Pos'].str.split().str.get(0) == selectedMainPos) &
                         (mainDf['Fin'] >= fin) & 
                         (mainDf['Hea'] >= hea) &
                         (mainDf['Pac'] >= pac) &
                         (mainDf['Pas'] >= pas) &
                         (mainDf['Dri'] >= dri) &
                         (mainDf['Pos'] >= pos) &
                         (mainDf['Han'] >= han) &
                         (mainDf['Ref'] >= ref)]

    #BAR CHART
    barchart = filteredData.hvplot.bar(x='Name', y='Max Value', xlabel='Player Name', ylabel='Max Value (Millions)', title=f'Players with Position: {position} and Max Value: <= {max_value}',
                                        rot=90, width=800, height=600, hover_cols=['Age', 'Division', 'Club', 'Nationality', 'Style', 'Preferred Foot', 'Best Role', 'Best Duty'], 
                                        flip_yaxis=False, responsive=True, legend=False)

    #STACKED BAR CHART
    p1 = filteredData.hvplot.bar(x='Name', y=['Fin','Hea','Pac','Pas','Dri','Pos','Han','Ref'], stacked=True, rot=90,
                                width=800, height=600, responsive=True, legend='top_left')
        
    return pn.Column(barchart, p1)
```

# PLEASE READ
### Run the code below to view the dashboard. When you run the dashboard becuase of the default settings the dashobard will show all Goalkeepers. Use the buttons and sliders to interact - you'll realise the graphs doesn't update in real time. After using the buttons and sliders close the tab and re-run the cell below again - now the graphs will be updated. Just so you know, there is a Wingback postion button - but no Wingbacks fit the conditions to be in the dataset, so if you select wingback, all graphs will be emtpy - we still implemeted for future improvements.

```python
scatter_plot, table_data = update_plot(max_value=valueSlider.value, position=posButton.value, fin=finSlider.value, hea=heaSlider.value, pac=pacSlider.value, pas=pasSlider.value, dri=driSlider.value, pos=posSlider.value, han=hanSlider.value, ref=refSlider.value)
bar_chart, line_graph = update_plot1(max_value=valueSlider.value, position=posButton.value, fin=finSlider.value, hea=heaSlider.value, pac=pacSlider.value, pas=pasSlider.value, dri=driSlider.value, pos=posSlider.value, han=hanSlider.value, ref=refSlider.value)

#LAYOUT

sidebar = [pn.pane.Markdown("# Find the player you're looking for"), 
             pn.pane.Markdown("#### Our Football Scouting Network Dashboard is a comprehensive tool designed to provide valuable insights into player data. Whether you're a coach, scout, or enthusiast, this dashboard offers a user-friendly interface to explore and analyse crucial data points. With intuitive sliders, interactive buttons, and dynamic graphs, you can delve deep into the world of football scouting effortlessly. Just hover over the graphs and extract data, yes its that easy - if needed, use the tools (on sidebar of some graphs) to reset, download etc"),    
             valueSlider, finSlider, heaSlider, pacSlider, pasSlider, driSlider, posSlider, hanSlider, refSlider, posButton]

scatterDisplay = pn.Column(
    scatter_plot
)

tableDisplay = pn.Column(
    pn.pane.Markdown("## All attributes and details"),
    table_data
)

barDisplay = pn.Column(
    bar_chart
)

lineDisplay = pn.Column(
    pn.pane.Markdown("## Attribute visualization"),
    line_graph
)

main = [pn.Row(scatterDisplay,tableDisplay),
       pn.Row(barDisplay, lineDisplay)]

template = pn.template.FastListTemplate(
    title='Football Scouting Network', 
    sidebar=sidebar,
    main=main,
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)


template.show()
```
